```python
#Importing necessary libraries for analysis

import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
df=pd.read_csv("../input/pakistans-largest-ecommerce-dataset/Pakistan Largest Ecommerce Dataset.csv")
df.head()
```

```python
#Checkin Columns in the dataframe
df.columns
```

```python
# filtering the required columns

df= df[['item_id', 'status', 'created_at', 'sku', 'price', 'qty_ordered',
       'grand_total', 'increment_id', 'category_name_1',
       'sales_commission_code', 'discount_amount', 'payment_method',
       'Working Date', 'BI Status', ' MV ', 'Year', 'Month', 'Customer Since',
       'M-Y', 'FY', 'Customer ID']]
df.head()
```

```python
# Converting into datetime

df['created_at'] = pd.to_datetime(df['created_at'])
df = df.sort_values(by = 'created_at')

df['Year'] = df['created_at'].dt.year
df['Month'] = df['created_at'].dt.month
df['Month_Name'] = df['created_at'].dt.month_name()
df.head()
```

```python
## Checking for null values
df.isna().sum()
```

```python
# Defining a date filter which we will use later
def df_filter_trend(year, xaxis, yaxis):
    df_year_filter=df[df['Year']== year]
    df_cat_sales= df_year_filter.groupby(xaxis)[yaxis].sum().reset_index()
    df_cat_sort= df_cat_sales.sort_values([xaxis], ascending=True)
    return df_cat_sort
    
```

```python
trend_2016= df_filter_trend(2016, "created_at", "grand_total")
trend_2017= df_filter_trend(2017, "created_at", "grand_total")
trend_2018= df_filter_trend(2018, "created_at", "grand_total")
```

```python
## Plotly Line chart to track the sales trend in each year

fig = make_subplots(rows=3, cols=1, subplot_titles=("Sales trend in Year 2016", "Sales trend in Year 2017",
                                                    "Sales trend in Year 2018"))

fig.append_trace(go.Scatter(
    x=trend_2016["created_at"],
    y=trend_2016["grand_total"],
    name="Sales in 2016",
), row=1, col=1)

fig.append_trace(go.Scatter(
    x=trend_2017["created_at"],
    y=trend_2017["grand_total"],
    name="Sales in 2017",
), row=2, col=1)

fig.append_trace(go.Scatter(
    x=trend_2018["created_at"],
    y=trend_2018["grand_total"],
    name="Sales in 2018",
), row=3, col=1)


fig.update_layout(height=800, width=900, title_text="Trackin the Sales Trend in each year")
fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=10),


)
fig.show()
```

```python
def df_filter_cat(year, xaxis, yaxis):
    df_year_filter=df[df['Year']== year]
    df_cat_sales= df_year_filter.groupby(xaxis)[yaxis].sum().reset_index()
    df_cat_sort= df_cat_sales.sort_values([yaxis], ascending=False)
    return df_cat_sort
```

```python
cat_2016= df_filter_cat(2016, "category_name_1", "grand_total")
cat_2017= df_filter_cat(2017, "category_name_1", "grand_total")
cat_2018= df_filter_cat(2018, "category_name_1", "grand_total")
```

```python
fig = make_subplots(rows=3, cols=1, subplot_titles=("Top Categories with Sales in Year 2016", "Top Categories with Sales in Year 2017",
                                                    "Top Categories with Sales in Year 2018"))

fig.append_trace(go.Bar(
    x=cat_2016["category_name_1"],
    y=cat_2016["grand_total"],
    name="2016"
), row=1, col=1)

fig.append_trace(go.Bar(
    x=cat_2017["category_name_1"],
    y=cat_2017["grand_total"],
    name="2017"
), row=2, col=1)

fig.append_trace(go.Bar(
    x=cat_2018["category_name_1"],
    y=cat_2018["grand_total"],
    name="2018"
), row=3, col=1)

fig.update_yaxes(title=dict(text= "Sales"))
fig.update_layout(height=800, width=900, title_text="Sales w.r.t Categories in each year")
fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=10),


)

fig.show()
```

```python
cat_2016= df_filter_cat(2016, "status", "grand_total")
cat_2017= df_filter_cat(2017, "status", "grand_total")
cat_2018= df_filter_cat(2018, "status", "grand_total")
```

```python
colors = ['rgb(177, 127, 38)', 'rgb(205, 152, 36)', 'rgb(99, 79, 37)',
                     'rgb(129, 180, 179)', 'rgb(124, 103, 37)']


specs = [[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}], [{'type':'domain'},{'type':'domain'}, {'type':'domain'}]]
fig = make_subplots(rows=2, cols=3, specs=specs, subplot_titles=['2016', '', '2017', '', 
                                                                 '2018'])

# Define pie charts
fig.add_trace(go.Pie(labels=cat_2016["status"], values=cat_2016["grand_total"], name='2016 Data',
                     marker_colors=colors), 1, 1)
fig.add_trace(go.Pie(labels=cat_2017["status"], values=cat_2017["grand_total"], name='2017 Data',
                     marker_colors=colors), 1, 3)
fig.add_trace(go.Pie(labels=cat_2018["status"], values=cat_2018["grand_total"], name='2018 Data',
                     marker_colors=colors), 2, 2)


# Tune layout and hover info
fig.update_traces(hoverinfo='label+percent+name+value', textinfo='label')
fig.update(layout_showlegend=False)

fig = go.Figure(fig)

fig.update_layout(title_text="Sales w.r.t Status in each Year", title_x=0.5),

fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=12),


)
fig.show()
```

```python
def df_top_10(year, xaxis, yaxis):
    df[xaxis]= df[xaxis].astype(str)
    df_year_filter=df[df['Year']== year]
    
    df_cat_sales= df_year_filter.groupby(xaxis)[yaxis].sum().reset_index()
    df_cat_sort= df_cat_sales.sort_values([yaxis], ascending=False).head(10)
    return df_cat_sort
```

```python
cust_2016_sales= df_top_10(2016, "Customer ID", "grand_total")
cust_2017_sales= df_top_10(2017, "Customer ID", "grand_total")
cust_2018_sales= df_top_10(2018, "Customer ID", "grand_total")

cust_2016_quantity= df_top_10(2016, "Customer ID", "qty_ordered")
cust_2017_quantity= df_top_10(2017, "Customer ID", "qty_ordered")
cust_2018_quantity= df_top_10(2018, "Customer ID", "qty_ordered")
```

```python

fig = make_subplots(rows=3, cols=2, horizontal_spacing = 0.3,vertical_spacing = 0.25,
                    subplot_titles=("Year 2016(Sales)", "Year 2016(Quantity)","Year 2017(Sales)", "Year 2017(Quantity)",
                                     "Year 2018(Sales)", "Year 2018(Quantity)"))

fig.append_trace(go.Bar(
    x=cust_2016_sales["Customer ID"],
    y=cust_2016_sales["grand_total"],
    name= "Top 10 Customer (Sales) in Year 2016"
), row=1, col=1)


fig.append_trace(go.Bar(
    x=cust_2016_quantity["Customer ID"],
    y=cust_2016_quantity["qty_ordered"],
     name= "Top 10 Customer (Quantity) in Year 2016"
), row=1, col=2)



fig.append_trace(go.Bar(
    x=cust_2017_sales["Customer ID"],
    y=cust_2017_sales["grand_total"],
     name= "Top 10 Customer (Sales) in Year 2017"
), row=2, col=1)


fig.append_trace(go.Bar(
    x=cust_2017_quantity["Customer ID"],
    y=cust_2017_quantity["qty_ordered"],
     name= "Top 10 Customer (Quantity) in Year 2017"
), row=2, col=2)



fig.append_trace(go.Bar(
    x=cust_2018_sales["Customer ID"],
    y=cust_2018_sales["grand_total"],
    name= "Top 10 Customer (Sales) in Year 2018"
), row=3, col=1)


fig.append_trace(go.Bar(
    x=cust_2018_quantity["Customer ID"],
    y=cust_2018_quantity["qty_ordered"],
     name= "Top 10 Customer (Quantity) in Year 2018"
), row=3, col=2)



fig.update_layout(height=800, width=900, title_text="Top 10 Customers w.r.t sales and quantity in each year")
fig.update_xaxes(title=dict(text= "Customer Number"))

fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=10),


)
fig.show()
```
